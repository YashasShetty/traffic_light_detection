```python
import cv2
from ultralytics import YOLO

# 1. Load the YOLOv10n model
# Ensure you have downloaded yolov10n.pt
model = YOLO("yolov10n.pt")

# 2. Run inference
# We filter for class 9 (traffic light) to keep the display clean
results = model.predict(source="images_to_check/2.jpeg", classes=[9])

# 3. Visualize results
for r in results:
    # plot() draws boxes and labels on the image
    annotated_frame = r.plot()
    
    # Display using OpenCV
    cv2.imshow("YOLOv10 Traffic Light Detection", annotated_frame)
    cv2.waitKey(0) # Press any key to close
    cv2.destroyAllWindows()
```
