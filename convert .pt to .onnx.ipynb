```python
from ultralytics import YOLO

# Load your trained model (or a pretrained one like yolo11n / yolo26n)
model = YOLO("yolo26m.pt") 

# Export to ONNX with dynamic axes
# 'dynamic=True' allows the model to accept various input shapes at runtime
model.export(
    format="onnx", 
    dynamic=True, 
    opset=19,      # Standard for maximum compatibility with OpenVINO/TensorRT
    simplify=True  # Removes redundant layers to boost speed
)
```
