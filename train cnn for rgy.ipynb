```python
import tensorflow as tf
from tensorflow.keras import layers, models,Input
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
```

```python
# --- 1. Define the Model ---
def build_traffic_classifier(input_shape=(64, 64, 3)):
    model = models.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(64, activation='relu'),
        layers.Dense(3, activation='softmax') 
    ])
    model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model
```

```python
# We flip the dimensions: (Height=128, Width=64, Channels=3)
def build_vertical_traffic_classifier(input_shape=(128, 64, 3)):
    model = models.Sequential([
        Input(shape=input_shape), 
        
        # Convolutional Block 1
        layers.Conv2D(32, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        
        # Convolutional Block 2
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        
        # Convolutional Block 3 (Optional: helps with taller features)
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.5), # Robustness layer
        layers.Dense(3, activation='softmax')
    ])
    
    model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model
```

```python
# Path to your Kaggle dataset (Update this to your actual path)
train_dir = 'archive/1/train'
val_dir = 'archive/1/val'
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=10,
    zoom_range=0.1
)
```

```python
val_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(128, 64),
    batch_size=32,
    class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(128, 64),
    batch_size=32,
    class_mode='categorical'
)
```

```python
model = build_vertical_traffic_classifier()
```

```python
history = model.fit(
    train_generator,
    epochs=10,
    validation_data=val_generator
)
```

```python
# --- 4. Save the Model ---
model.save('traffic_light_cnn_vertical.h5')
print("Model saved as traffic_light_cnn.h5")
```
