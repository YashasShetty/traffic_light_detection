```python
import os
import subprocess

def fast_stitch_dashcam(input_folder, output_filename):
    # 1. Get and sort your files
    video_files = sorted([f for f in os.listdir(input_folder) if f.endswith('.MP4')])
    
    if not video_files:
        print("No videos found.")
        return

    # 2. Create a temporary 'input_list.txt' file that FFmpeg needs
    # Format: file 'path/to/video.mp4'
    list_file = "inputs.txt"
    with open(list_file, "w") as f:
        for filename in video_files:
            # We use absolute paths to avoid issues with FFmpeg finding the files
            abs_path = os.path.abspath(os.path.join(input_folder, filename))
            f.write(f"file '{abs_path}'\n")

    # 3. Run FFmpeg via subprocess
    # -f concat: use the concat demuxer
    # -safe 0: allow any file path
    # -c copy: THE MAGIC. This copies the stream without re-encoding.
    command = [
        "ffmpeg", "-y", "-f", "concat", "-safe", "0", 
        "-i", list_file, "-c", "copy", output_filename
    ]

    print("Stitching via Stream Copy (this should be very fast)...")
    try:
        subprocess.run(command, check=True)
        print(f"Done! Saved to {output_filename}")
    except subprocess.CalledProcessError as e:
        print(f"Error: Make sure FFmpeg is installed on your system. {e}")
    finally:
        # 4. Clean up the temp file
        if os.path.exists(list_file):
            os.remove(list_file)

# Usage
fast_stitch_dashcam("./video/CRV_front_cam_28_feb", "./video_temp/28feb_crv.mp4")
```
