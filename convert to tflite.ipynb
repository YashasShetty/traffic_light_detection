```python
from ultralytics import YOLO

# Load your trained YOLO26m model
model = YOLO("yolo26m.pt")

# Export to TFLite format
model.export(format="tflite") 
# Result: 'yolo26m_float32.tflite'
```

```python
from ultralytics import YOLO

model = YOLO("yolo26m.pt")

# MANDATORY: imgsz must be 320 or 640. 
# MANDATORY: nms=False (The plugin handles NMS in Dart/Native code)
model.export(
    format="tflite", 
    imgsz=320,      # Smaller is often more stable for the Flutter plugin
    int8=False,     # Start with float32 to rule out quantization errors
    nms=False       # IMPORTANT: Do not include NMS in the TFLite graph
)
```
