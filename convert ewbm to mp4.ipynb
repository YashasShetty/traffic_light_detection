```python
import subprocess

def convert_webm_to_mp4(input_file, output_file):
    command = [
        'ffmpeg',
        '-i', input_file,         # Input WebM
        '-c:v', 'libx264',        # Convert to H.264 (Highly compatible)
        '-crf', '23',             # Quality setting (18-28 is good, lower is better)
        '-preset', 'medium',      # Encoding speed vs compression
        '-c:a', 'aac',            # Convert audio to AAC
        '-b:a', '128k', 
        output_file
    ]
    
    try:
        subprocess.run(command, check=True)
        print(f"Conversion successful: {output_file}")
    except subprocess.CalledProcessError as e:
        print(f"Error during conversion: {e}")

# Usage
convert_webm_to_mp4('video/2.webm', 'video/3.mp4')
```

```python
import subprocess

def convert_webm_to_mp4_gpu(input_file, output_file):
    command = [
        'ffmpeg',
        '-hwaccel', 'cuda',           # Use CUDA to help decode the input
        '-i', input_file, 
        '-c:v', 'h264_nvenc',         # The NVIDIA Hardware Encoder
        '-preset', 'p4',              # Performance preset (p1=fastest, p7=slowest/best)
        '-rc', 'vbr',                 # Variable bit rate
        '-cq', '24',                  # Quality (Constant Quality mode)
        '-c:a', 'aac', 
        output_file,
        '-y'                          # Overwrite if exists
    ]
    
    try:
        print("Starting GPU-accelerated conversion...")
        subprocess.run(command, check=True)
        print(f"Done! Saved to: {output_file}")
    except subprocess.CalledProcessError as e:
        print(f"GPU conversion failed. Error: {e}")
        print("Check if NVIDIA drivers and FFmpeg with NVENC support are installed.")

# Run it
convert_webm_to_mp4_gpu('video/2.webm', 'video/3.mp4')
```
