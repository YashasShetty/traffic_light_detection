```python
from pydub import AudioSegment

# 1. Load the m4a file
# pydub uses ffmpeg behind the scenes to decode the m4a
audio = AudioSegment.from_file("music/unknown.m4a", format="m4a")

# 2. Export as ogg
# 'libvorbis' is the standard codec for ogg files
audio.export("music/unknown.ogg", format="ogg", codec="libvorbis")

print("Conversion complete: track.ogg is ready for Pygame!")
```
