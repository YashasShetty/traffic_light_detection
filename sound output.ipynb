```python
import pygame

class SoundManager:
    def __init__(self):
        pygame.mixer.init()
        self.volume = 0.5  # Default volume
        # Load your sounds into a dictionary for easy access
        self.sounds = {
            "hollow": pygame.mixer.Sound("music/hollow.mp3"),
            # "click": pygame.mixer.Sound("music/click.wav")
        }
        self.update_all_volumes()

    def update_all_volumes(self):
        """Applies the current volume setting to all loaded sounds."""
        for sound in self.sounds.values():
            sound.set_volume(self.volume)

    def play(self, sound_name):
        if sound_name in self.sounds:
            self.sounds[sound_name].play()

    def set_master_volume(self, level):
        """Sets volume from 0.0 to 1.0"""
        self.volume = max(0.0, min(1.0, level)) # Clamps value between 0 and 1
        self.update_all_volumes()
        print(f"Volume set to: {int(self.volume * 100)}%")

# --- How to use it ---
audio = SoundManager()

# Event happens:
audio.play("hollow")

# User turns down the volume:
audio.set_master_volume(0.2)
audio.play("hollow") # Now plays much quieter
```

```python
#
import pygame
import time

# 1. INITIALIZE ONCE (Globally or in your main startup function)
pygame.mixer.init()

# 2. PRE-LOAD YOUR SOUNDS
# Loading files from disk is slow; doing it once at the start makes playback instant.
HOLLOW_SOUND = pygame.mixer.Sound("music/hollow.mp3")

def trigger_event_sound():
    """Call this function whenever your event happens."""
    # .play() on a Sound object is non-blocking (the script keeps running)
    HOLLOW_SOUND.play()
    print("Event sound triggered!")

# --- Simulation of your app ---
try:
    print("App is running... Press Ctrl+C to stop.")
    while True:
        # Simulate an event happening
        trigger_event_sound()
        
        # Wait 5 seconds before the next event
        time.sleep(5) 

except KeyboardInterrupt:
    print("\nShutting down...")
finally:
    # 3. CLEAN UP ONCE (Only when the entire program closes)
    pygame.mixer.quit()
```
